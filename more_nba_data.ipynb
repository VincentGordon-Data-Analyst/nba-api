```python
import pandas as pd

games_df = pd.read_csv('./kaggle_nba_files/games.csv')
games_df
```

# Game Information

```python
games_df.info()
```

```python
# Find total number of null values from each column
games_df.isnull().sum()
```

```python
# Check the years of the dataset
games_df['SEASON'].unique()
```

```python
# Find rows with Null values
games_df[games_df['PTS_home'].isnull()] 
```

```python
# Take only games from specified date then drop rows with any null values
cleaned_games_df = games_df[games_df['GAME_DATE_EST'] >= '2019-01-01'].dropna()

# Check the first few rows of the cleaned Dataframe
cleaned_games_df.head()
```

```python
# Look at column names
cleaned_games_df.columns.values
```

```python
# Rename columns 
cleaned_games_df.rename(columns={'GAME_DATE_EST': 'date',
                                 'TEAM_ID_home': 'home', 
                                 'PTS_home': 'home_points', 
                                 'TEAM_ID_away': 'visitor', 
                                 'PTS_away': 'visitor_points'}, inplace=True)

print(cleaned_games_df)
```
