```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import numpy as np

# to ignore the warnings
from warnings import filterwarnings
```

```python
df = pd.read_csv('./data/nbaFeaturedEngineered.csv')
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
# Create first target column 'winner'
df['home_win'] = df.apply(
    lambda row: 1 if row['home_points'] > row['visitor_points'] else 0,
    axis=1
)
# Create second target column 'total_points'
df['total_points'] = df['home_points'] + df['visitor_points']

# Display
df.head()
```

```python
# Apply one-hot encoding and convert to booleans to integers
encoded_df = pd.get_dummies(df, columns=['visitor_team', 'home_team'])

# Display
new_df = encoded_df.replace({True: 1, False: 0})
new_df.head()
```

# Machine Learning

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor

# Split Data
X = new_df.drop(columns=['home_win', 'total_points', 'date'])
y_winner = df['home_win']
y_total = df['total_points']

X_train, X_test, y_winner_train, y_winner_test = train_test_split(X, y_winner, test_size=0.2, random_state=42)
_, _, y_total_train, y_total_test = train_test_split(X, y_total, test_size=0.2, random_state=42)

# Train Models
clf = RandomForestClassifier()
clf_model = clf.fit(X_train, y_winner_train)

reg = RandomForestRegressor()
reg_model = reg.fit(X_train, y_total_train)

# Predictions
winner_prediction = clf_model.predict(X_test)
total_predictions = reg_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_winner_pred = clf_model.predict(X_test)
print(confusion_matrix(y_winner_test, y_winner_pred))
print(classification_report(y_winner_test, y_winner_pred))
print("Accuracy: ", accuracy_score(y_winner_test, y_winner_pred))
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_total_pred = reg_model.predict(X_test)
print("MSE: ", mean_squared_error(y_total_test, y_total_pred))
print("R2 Score: ", r2_score(y_total_test, y_total_pred))
```

```python
# Input Teams
visitor_team = input("Enter Away Team: ")
home_team = input("Enter Home Team: ")

# Prepare Input Data
input_data = {col: 0 for col in X.columns}  # Initialize all columns with 0

# Set team-specific features
if f'home_team_{home_team}' in input_data:
    input_data[f'home_team_{home_team}'] = 1
if f'visitor_team_{visitor_team}' in input_data:
    input_data[f'visitor_team_{visitor_team}'] = 1

# For numeric features, use the mean (or other default values)
for col in X.select_dtypes(include=['float64', 'int64']).columns:
    input_data[col] = X[col].mean()

# Convert to DataFrame
input_df = pd.DataFrame([input_data])

# Predict Winner
predicted_winner = clf_model.predict(input_df)[0]
winning_team = home_team if predicted_winner == 1 else visitor_team

# Predict Total Points
predicted_total_points = reg_model.predict(input_df)[0]

# Display Results
print("Predicted Winner:", winning_team)
```
